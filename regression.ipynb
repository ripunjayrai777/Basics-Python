```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv("data\htwgtMale.csv")
```

```python
X = df[["Height"]].values
Y = df["Weight"].values
```

```python
X,Y
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, random_state=48, test_size=0.3)
```

```python
X_train.shape,X_test.shape, Y_train.shape, Y_test.shape
```

```python
plt.plot(X, Y, c='g', marker="^", label='Original Data')
plt.xlabel('Height(cm)')
plt.ylabel('Weight(Kg)')
plt.show( )
```

```python
lr=LinearRegression()
#Training of Data
lr.fit(X_train, Y_train )
#Prediction
Y_pred=lr.predict(X_test)
print('Shape of y_pred is:')
print(Y_pred.shape)

print("Test Val<--Vs-->Predicted Val")
for i in range(len(Y_test)):
   print(Y_test[i],"<--Vs-->",Y_pred[i])
```

```python
lr.intercept_
```

```python
m=lr.coef_
c=lr.intercept_
print('Slope =',m,'y-intercept=',c)

yCalc= m*X+c
plt.plot(X,yCalc)
    
plt.scatter(X_test, Y_pred, c='r', marker="s", label='Prediction')
plt.scatter(X_test, Y_test, c='b', marker="o", label='Test Case')
plt.legend(loc='upper left')
plt.xlabel('Height(cm)')
plt.ylabel('Weight(Kg)') 
plt.show( )
```
