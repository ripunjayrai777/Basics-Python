```python
import pandas as pd
```

```python
dict = {
    'First Score':  [100, 90, None, 95],
    'Second Score': [30, 45, 56, None],
    'Third Score':  [None, 40, 80, 98]
}
```

```python
df = pd.DataFrame(dict)
```

```python
print(df)
```

```python

print(df.isnull())
```

```python
print(df.notnull())
```

```python
print("Original DataFrame:")
print(df)
```

```python
df_filled = df.fillna(method='pad')
print(df_filled)
```

```python
df_filled2 = df.ffill()
print(df_filled2)
```

```python
df_filled2 = df.bfill()
print(df_filled2)
```

```python
df_filled4 = df.fillna(method='bfill')
print(df_filled4)
```

```python
df_filled5 = df.fillna(method='ffill')
print(df_filled5)
```
