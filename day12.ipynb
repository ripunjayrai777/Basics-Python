Filter Data Using Pandas

```python
import pandas as pd
```

```python

data = {
    'name': ['ripu', 'ayush', 'nadeem', 'sajid', 'Abhay'],
    'age': [15, 34, 10, 20, 34],
    'salary': [50000, 78000, 45000, 78000, 80000]
}
df = pd.DataFrame(data)
```

result = df[df['age']==10 ]
print(result)

```python
result = df[df['age'] > 10]
print(result)
```

```python
result3 = df[df['name']=='ayush']
print(result3)
```

Example 2

```python
result2 = df[df['salary'] > 50000]
print(result2)
```

Now, with multiple conditions...

```python
result3 = df[(df['age']>20) & (df['salary']>50000)]
print(result3)
```
