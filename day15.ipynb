```python
a = np.array( [1.1,  2.5,  0.6, 4.7, 5.3] ) 

print('Base of Array-A: ', a.base )
```

```python
b=a.copy()
b[0]=10
print('Values of Array -A ')

print(b)
print(a)
print('Base of Array-B: ', a.base )
```

```python
a = np.array( [1.1,  2.5,  0.6, 4.7, 5.3] ) 
c=a.view()
c[0]=10;
print(c)
print(a)
print('Base of Array-C: ', c.base )
```
