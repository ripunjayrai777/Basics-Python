```python
import matplotlib.pyplot as plt
subjects = ('Python', 'C++', 'Java', 'Perl', 'Scala', 'Lisp')
performance = [10, 8, 6, 4, 2, 1]
plt.grid(color='g', linestyle='--', linewidth=2, axis='y', alpha=0.3)
plt.bar(subjects, performance, align='center', alpha=0.5)
plt.ylabel('Usage')
plt.title('Programming language usage')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

data1 = [23,85, 72, 43, 52]
data2 = [42, 35, 21, 16, 9]
width =0.3
X1= np.arange(len(data1))
X2= np.arange(len(data2))+ width
plt.bar(X1, data1, color='r',width=width, label='Boys', align='center')
plt.bar(X2, data2, color='b', width=width, label='Girls', align='center')
plt.legend()
plt.show()
```

```python

# Subjects
subjects = ['Math', 'Physics', 'Ch0emistry', 'English', 'CS']

# Marks
boys_marks = [75, 80, 72, 68, 85]
girls_marks = [78, 85, 70, 75, 90]


# Bar width
width = 0.35

# X positions (without numpy)
x = list(range(len(subjects)))

# Shifted positions for girls
x2 = [i + width for i in x]

# Grid
plt.grid(axis='y', linestyle='--', alpha=0.4)

# Boys bars
plt.bar(x, boys_marks, width=width, color='blue', align='center', label='Boys')

# Girls bars
plt.bar(x2, girls_marks, width=width, color='pink', align='center', label='Girls')

# Labels & Title
plt.ylabel("Marks")
plt.title("Class Performance Comparison")

# Legend
plt.legend()

# Show plot
plt.show()
```

```python

data1 = [23,85,72,43,52]
data2 = [42,35,21,16,9]
labels = ["Eng", "Maths", "Hindi", "Comp", "Science"]  #check
x = range(len(data1))

plt.bar(x, data1)
plt.bar(x, data2, bottom=data1)
plt.xticks(x, labels )  # use for labels on x-axis
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# data to plot
n_groups = 4
Sales_Voltas = (90, 55, 40, 65)
Sales_Samsung = (85, 62, 54, 20)
Year_Sales= ('2016', '2017', '2018', '2020')

# create plot
y_pos= np.arange(n_groups)
bar_width = 0.35
opacity = 0.8

rects1 = plt.bar(y_pos , Sales_Voltas , bar_width, alpha=opacity, color='b', label='Voltas')

rects2 = plt.bar(y_pos + bar_width, Sales_Samsung ,  bar_width, alpha=opacity, color='g', label='Samsung')

plt.xlabel('Year')
plt.ylabel('Sales')
plt.title('Sales Scores')
plt.xticks(y_pos + bar_width,  Year_Sales)
plt.legend()

plt.show()
```
