```python
import numpy as np

a = np.array([[1, 2, 3, 4],
              [5, 6, 7, 8]])
```

Simple Nested Loop – Print Each Value in row

```python
print('Simple Loop - Print each value in row')

for row in a:
    for col in row:
        print(col)
```

np.nditer() – Print Each Value in row with special iterator

```python

for val in np.nditer(a):
    print(val)
```

```python
for id, val in np.ndenumerate(a):
    print(id, val)
```

Sorting Array

```python
b = np.array([1,4,3,2,5,8,7,6])
print(b.shape)

print(np.sort(b))
```

Searching in Array

```python
print("Search the Even Numbers")
x = np.where(b % 2 == 0)
print(x)
```

Important: np.where() value nahi, index return karta hai.

```python
print("Search the Even Numbers")
abc = (b % 2 == 0)
print(abc)
```

```python
ind = (b % 2 == 0)
arr = b[ind]
print(arr)
```
