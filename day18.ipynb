ylabel() - use for label the Y- axis

```python
#25 scatter plot
import matplotlib.pyplot as plt

# Data
study_hours = [1, 2, 3, 4, 5]
marks = [40, 50, 65, 70, 85]

# Create Scatter Plot
# plt.scatter()
plt.plot(study_hours, marks, marker="o")

# Labels and Title
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("Study Hours vs Marks")

# Show plot
plt.show()
```
