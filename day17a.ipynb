```python
# Data
x = [1,2,3,4]
y1 = [1,7,9,16]
y2 = [2,3,5,7]
y3 = [2,3,5,6]
y4 = [2,7,4,7]

# First subplot
plt.subplot(2,1,1)   # 2 rows, 1 column, position 1
plt.plot(x, y1, 'r')
plt.title("Square Numbers")

# Second subplot
plt.subplot(2,1,1)   # 2 rows, 1 column, position 2
plt.plot(x, y2, 'b--')
plt.title("Prime Numbers")
plt.tight_layout()  
```

```python
#some of case we can write like--
plt.subplot(221)
#first plot
plt.plot(x, y1, 'r-')
#second plot
plt.subplot(222)
plt.plot(x, y2, 'b--')
#third plot
plt.subplot(223)
plt.plot(x, y3, 'g-.')
```
